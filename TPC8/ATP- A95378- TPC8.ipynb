```python
# Leitura/carregamento da informação do ficheiro

def convBool(v):
    return v=='true'

def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline()
    for linha in f: 
        novalinha=linha.strip('\n')
        campos=novalinha.split(",")
        emd=[]
        emd.append("emd" + str(campos[1]))
        emd.append(campos[2])
        emd.append(campos[4]+ " " +campos[3])
        emd = emd + campos[5:]
        emd[-1]= convBool(emd[-1])
        emd[-2]= convBool(emd[-2])
        bd.append(emd)   
    return bd
```

```python
BD = lerDataset("emd.csv")
print(BD)
```

```python
#listar dataset
def listarDataset(bd):
    bd.sort(reverse=True, key=lambda emd: emd[1]) 
    print("ID     | NOME        | DATA      | RESULTADO")
    print("--------------------------------------------------------")
    for exame in bd:
        if exame[-1]:
            res='Apto'
        else:
            res='Não Apto'
        print(exame[0]  +" | " + exame[2] + "  | "  +  exame[1] +  " | " +  res)

listarDataset(BD)    
```

```python
# Consulta de um registo
def consultarDataset(bd, id):
    for exame in bd:
        if exame[0]==id:
            if exame[-1]:
                res='Apto'
            else:
                res='Não Apto'
            print(exame[0]  +" | " + exame[2] + "  | "  +  exame[1] +  " | " +  res)

```

```python
#lista de modalidades ordenada alfabeticamente e sem repetições
def modalidades(bd):
    modalidades=[]
    for exame in bd:
        if exame[6] not in modalidades:
            modalidades.append(exame[6]) 
    modalidades.sort()
    return modalidades

modalidades(BD)
```

```python
#lista de pares indicando quantos EMD estão registados em cada modalidade
def distribPorModalidade(bd):
    distmodalidade={}
    for m in bd:
        if m[6] in distmodalidade.keys():
             distmodalidade[m[6]]=distmodalidade[m[6]] + 1
        else:
            distmodalidade[m[6]]=1
            
    return distmodalidade

distribPorModalidade(BD)
```

```python
#lista de pares indicando quantos EMD estão registados por cada clube
def distribPorClube(bd):
    distclube={}
    for c in bd:
        if c[7] in distclube.keys():
             distclube[c[7]]=distclube[c[7]] + 1
        else:
            distclube[c[7]]=1
            
    return distclube

distribPorClube(BD)
```

```python
# lista de pares indicando quantos EMD estão registados por cada ano
def distribPorAno(bd):
    distano={}
    for exame in bd:
        ano=exame[1][0:4]
        if ano in distano.keys():
            distano[ano]= distano[ano]+1
        else:
            distano[ano]= 1
    return distano

distribPorAno(BD)
```

```python
#distribuição por qualquer um dos campos da BD
def distrib(bd, campo):
    dist={}
    for exame in bd:
        if exame[campo] in dist.keys():
             dist[exame[campo]]=dist[exame[campo]] + 1
        else:
            dist[exame[campo]]=1
            
    return dist
```

```python
import matplotlib.pyplot as plt
```

```python
def plotDistribPorModalidade(bd):
    plt.rcParams["figure.figsize"] = (16,8)
    
    distribuicao=distribPorModalidade(bd)

    x = list(range(1,len(distribuicao.keys())+1))
    y = list(distribuicao.values())

    modalidades = list(distribuicao.keys())

    plt.bar(x, y, tick_label = modalidades , width = 0.8, color = ['turquoise'])

    plt.xlabel('Modalidade')
    plt.ylabel('Nº de alunos')
    plt.title('Distribuição por modalidade')
    plt.xticks(rotation=90)
 

    plt.show()
    
plotDistribPorModalidade(BD)
```

```python
def plotDistribPorAno(bd):
    plt.rcParams["figure.figsize"] = (16,8)
    
    distribuicao=distribPorAno(bd)

    x = list(range(1,len(distribuicao.keys())+1))
    y = list(distribuicao.values())

    ano = list(distribuicao.keys())

    plt.bar(x, y, tick_label = ano , width = 0.8, color = ['skyblue'])

    plt.xlabel('ano')
    plt.ylabel('Nº de alunos')
    plt.title('Distribuição por ano')
    plt.xticks(rotation=90)
 

    plt.show()
    
plotDistribPorAno(BD)
```

```python
#gráfico com a distribuição passada como argumento
def plotDistrib(bd, campo):
    plt.rcParams["figure.figsize"] = (16,8)
    
    distribuicao=distrib(bd, campo)

    x = list(range(1,len(distribuicao.keys())+1))
    y = list(distribuicao.values())

    campo = list(distribuicao.keys())

    plt.bar(x, y, tick_label = campo , width = 0.8, color = ['skyblue'])

    plt.ylabel('Nº de alunos')
    plt.title('Distribuição')
    plt.xticks(rotation=90)
 

    plt.show()
    
```
