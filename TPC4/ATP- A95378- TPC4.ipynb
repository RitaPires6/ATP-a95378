```python
import random
lista=[]

def criarfracao():
    numerador=int(input("Diga o numerador:"))
    denominador=int(input("Diga o denominador:"))
    return (numerador,denominador)

def verfracao(f):
    n,d=f
    print(str(n)+"/"+str(d))
    
def mdc(a,b):
    if a<b:
        return mdc(b,a)
    elif a%b==0:
        return b
    else:
        return mdc(a,a%b)
    
def simplificarfracao(f):
    n,d=f
    m=mdc(n,d)
    return(int(n/m),int(d/m))

def somarfracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificarfracao((n1*d2+n2*d1, d2*d1))

def subtrairfracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificarfracao((n1*d2-n2*d1, d2*d1))

def multiplicarfracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificarfracao((n1*n2, d1*d2))

def dividirfracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificarfracao((n1*d2, n2*d1))

def criarlista():
    t=int(input("diga qual a dimensao da lista que pretende gerar: "))
    listafrac=[]
    for i in range(1,t):
        n=random.randrange(1,100)
        d=random.randrange(1,100)
        f=(n,d)
        listafrac.append(f)
    return listafrac
    print(listafrac)

    
def escolherlista():
    x=(input("se quer uma lista aleatoria escreva, aleatoria, se quer uma lista criada por si escreva, nova, ou se quer uma lista de todas as frações já criadas e dos resultados dos calculos efetuados escreva, anteriores : "))
    if x=="aleatoria":
        L=criarlista()
        return L
        
    elif x=="nova":
        L=[]
        t1=int(input("diga a dimensão da sua lista: "))
        for i in range(t1):
            n=int(input("Diga o numerador:"))
            d=int(input("Diga o denominador:"))
            f=(n,d)
            L.append(f)
        return L
        
    elif x=="anteriores":
        L=lista
        return L
    
    
def maiorfracao(L):
    print(L)
    max=L[0]
    for i in range(1,len(L)):
        n1,d1=max
        n2,d2=L[i]
        if(n1*d2)<(n2*d1):
            max=L[i]        
    return max

def somarlista(L):
    print(L)
    soma=L[0]
    for i in range (1, len(L)):
        soma=somarfracao(soma, L[i])
    return simplificarfracao(soma)

def menu():
    print("""
Menu:
1) Criar Fração
2) Simplificar a Fração
3) Somar Frações
4) Subtrair Frações
5) Multiplicar Frações
6) Dividir Frações
7) Gerar lista aleatoria de n Frações
8) Somar as Frações de uma Lista
9) Maior Fração de uma Lista
0) Sair""")
    opcao=10
    while opcao!=0:
        opcao=int(input("insira a opçao quer usar: "))
        
        if opcao==1:
            ans=0
            ans=criarfracao()
            lista.append(ans)
            print(verfracao(ans))
            
            
        elif opcao==2:
            print(verfracao(simplificarfracao(ans)))
            
            
        elif opcao==3:
            f1=ans
            print("a sua primeira fracao ja esta criada: ")
            print( verfracao(simplificarfracao(ans)))
            f2=criarfracao()
            lista.append(f2)
            print(verfracao(f2))
            ans=(simplificarfracao(somarfracao(f1, f2)))
            lista.append(ans)
            print(verfracao(ans))
            
        
        elif opcao==4:
            f1=ans
            print("a sua primeira fracao ja esta criada: ")
            print(verfracao(simplificarfracao(ans)))
            f2=criarfracao()
            lista.append(f2)
            print(verfracao(f2))
            ans=(simplificarfracao(subtrairfracao(f1, f2)))
            lista.append(ans)
            print(verfracao(ans))
          
          
        elif opcao==5:
            f1=ans
            print("a sua primeira fracao ja esta criada: ")
            print(verfracao(simplificarfracao(ans)))
            f2=criarfracao()
            lista.append(f2)
            print(verfracao(f2))
            ans=(simplificarfracao(multiplicarfracao(f1, f2)))
            lista.append(ans)
            print(verfracao(ans))
    
        elif opcao==6:
            f1=ans
            print("a sua primeira fracao ja esta criada: ")
            print(verfracao(simplificarfracao(ans)))
            f2=criarfracao()
            lista.append(f2)
            print(verfracao(f2))
            ans=(simplificarfracao(dividirfracao(f1, f2)))
            lista.append(ans)
            print(verfracao(ans))
            
            
        elif opcao==7:
            print(criarlista())
            
        elif opcao==8:
            ans=(somarlista(escolherlista()))
            lista.append(ans)
            print(verfracao(ans))
           
        elif opcao==9:
            print(verfracao(maiorfracao(escolherlista())))
            
        elif opcao==0:
            print("OFF")
            
        else:
            print("invalida")
            
menu()
            
        
     
```
